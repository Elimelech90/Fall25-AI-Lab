4. Data Preprocessing 

4.1. Manage null values

```python
# df.drop('id', axis=1, inplace=True)
df = df.drop('id', axis=1)
```

```python
# df.drop('City', axis=1, inplace=True)
df = df.drop('City', axis=1)
```

```python
# Mean, Median, Mode

# df['CGPA'].mean()
# df['CGPA'].median()
df['CGPA'].mode()[0]
```

```python
# df['ID'].fillna(df['ID'].mode()[0], inplace=True)
df['Sleep Duration'] = df['Sleep Duration'].fillna(df['Sleep Duration'].mode()[0])
```

```python
# For Object / String Type
def fillnaObjectMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['Sleep Duration']
fillnaObjectMode(columns)
```

```python
# For Int Type
def fillnaIntMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['Work/Study Hours', 'Age']
fillnaIntMode(columns)
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i: df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))
        df[i] = data_LE.transform(df[i])

columns = ['Gender', 'Dietary Habits']
convertObjToInt(columns)
```
